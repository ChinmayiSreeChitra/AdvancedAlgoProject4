```python
# Finding Optimal Value using library

from pulp import LpProblem, LpMaximize, LpVariable, LpMinimize
import numpy as np
import matplotlib.pyplot as plt
#optimal value calculation
# Create a maximization problem
primal = LpProblem("Primal", LpMaximize)
# Define variables
x1 = LpVariable("x1", lowBound=0)
x2 = LpVariable("x2", lowBound=0)

# Define the objective function
primal += 3 * x1 + 2 * x2

# Define constraints
primal += 2 * x1 + 1 * x2 <= 10
primal += 1 * x1 + 2 * x2 <= 7

# Solve the primal problem
primal.solve()

# Retrieve the optimal values
optimal_x1 = x1.varValue
optimal_x2 = x2.varValue
optimal_objective = primal.objective.value()

print("Optimal x1:", optimal_x1)
print("Optimal x2:", optimal_x2)
print("Optimal Objective Value:", optimal_objective)
```

```python
# Finding optimal value from graph
import numpy as np
import matplotlib.pyplot as plt

# Define the constraints
x = np.linspace(0, 20, 400)
constraint1 = (10 - 2 * x) / 1
constraint2 = (7 - 1 * x) / 2

# Define the objective function
objective = (3 * x) / 2

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1, label='2x + y ≤ 10')
plt.plot(x, constraint2, label='x + 2y ≤ 7')
plt.plot(x, objective, label='3x + 2y​', linestyle='--')

# Fill the feasible region
plt.fill_between(x, np.minimum(constraint1, constraint2), where=(x >= 0) & (x <= 20), alpha=0.3)

# Add labels and legends
plt.xlabel('x1')
plt.ylabel('x2')
plt.xlim(0, 20)
plt.ylim(0, 20)
plt.legend(loc='upper right')

# # Mark the optimal point
# optimal_x1 = 10
# optimal_x2 = 10
plt.scatter(optimal_x1, optimal_x2, color='red', marker='o', label=f'Optimal Point({optimal_x1}, {optimal_x2})')

# Show the plot
plt.title('Graphical Solution of Linear Programming')
plt.grid(True)
plt.legend()
plt.show()
```
